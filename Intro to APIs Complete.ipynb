```python
def get_racer_info(data):
    new_data=[]
    for racer in data:
        racer_dict={}
        racer_name = f"{racer['Driver']['givenName']} {racer['Driver']['familyName']}"
        
        racer_dict[racer_name]={
            'first_name':racer['Driver']['givenName'],
            'last_name':racer['Driver']['familyName'],
            'position':racer['position'],
            'wins':racer['wins'],
            'DOB':racer['Driver']['dateOfBirth'],
            'nationality':racer['Driver']['nationality'],
            'contructor':racer['Constructors'][0]['name']
        }
        new_data.append(racer_dict)
    return new_data

get_racer_info(data)
```

```python
def get_driver_info_by_year_and_round(year,r):
    url = f"http://ergast.com/api/f1/{year}/{r}/driverStandings.json"
    response = requests.get(url)
#     if not response.ok:
    if not response.json()["MRData"]["StandingsTable"]["StandingsLists"]:
        return "We had an error loading your data likely the year or round is not in the database"
    data = response.json()["MRData"]["StandingsTable"]["StandingsLists"][0]["DriverStandings"]
    return get_racer_info(data)

get_driver_info_by_year_and_round(2022,4)
```

check out this Pokemon API https://pokeapi.co/
Use the requests package to connect to this API and get and store data for 5 different pokemon.
Get the pokemons: name, atleast one ability's name, base_experience, and the URL for its sprite (an image that shows up on screen) for the 'front_shiny'

```python
import requests, json

def chooseyourpokemon():
    Pokemon = []
    i = 0
    for i in Pokemon:
        if i < 6:
            user_choice = input('Please enter the name of the Pokemon you would like on your team or enter "Q" to cancel.')
            if user_choice != 'Q':
                pokemon_apicall = f'https://pokeapi.co/api/v2/pokemon/{user_choice.lower()}'
                pokemon_apicall_response = requests.get(pokemon_apicall)
                if pokemon_apicall_response.ok == True:
                    pokemon_data = {'Name': pokemon_apicall_response.json()['forms'][0]['name'],
                                    'Ability': pokemon_apicall_response.json()['abilities'][0]['ability']['name'],
                                    'Base Experience': pokemon_apicall_response.json()['base_experience'],
                                    'Sprite': pokemon_apicall_response.json()['sprites']['front_shiny']}
                    Pokemon.append(pokemon_data)
                    print(Pokemon)
                    i += 1
                else:
                    print(f'Please enter a valid Pokemon name or ID')
            else:
                break
        else:
            print(f'Your party is currently full')
            print(Pokemon)
    return Pokemon

chooseyourpokemon()
```
